# 기본 그래프 그리기

```python
temp = [10.8, 8.5, 8.7, 7.9, 5.1,
        8.1, 8.2, 10.4, 11, 9.9,
        7.8, 9.5, 11.6, 8.2, 7.7,
        6.6, 11.1, 12.2, 12.6, 10.4, 
         13.7, 14.1, 12.6, 11.5, 13,
        14.5, 13.7, 9.4, 10.4, 11.6, 12.9]
```

#그래프 제목 추가하기

```python
plt.title('3월 기온')
plt.xlabel('날짜')
plt.ylabel('기온')

plt.plot(temp)
plt.show()
```

### 선 모양 정하기
+ **색상**
 - b(파란색), g(초록색), r(빨간색), c(청록색), y(노란색), k(검은색), w(흰색)
+ **마커**
 - o(원), v(역삼각형), ^(삼각형), s(네모), +(플러스), .(점)
+ **선스타일**
 - \\-(solid line), --(dashed line ), -. (dash-dot line), : (dotted line )

```python
plt.plot(temp, 'go--')
plt.show()
```

### 그래프 크기 조절
+ plt.figure()
 - figsize를 이용하여 가로, 세로 길이 조절 가능 (inch 단위)

```python
plt.figure(figsize=(10,6))

plt.title('3월 기온')
plt.xlabel('날짜')
plt.ylabel('기온')

plt.plot(temp, 'go--')
plt.show()
```

### 격자 및 수평/수직선 추가하기
#### 그리드 (grid)
+ grid() 함수의 첫번째 파라미터를 True로 설정하면 그래프에 그리드 (grid)가 표시
+ alpha:  그리드의 투명도를 설정
 - 0으로 설정하면 투명하게, 1은 불투명하게 표시
+ linestyle : 선의 스타일을 대쉬 (Dashed)로 설정

#### 수평 / 수직선 그리기
+ plt.axvline(x, color)
 - 축을 가로지르는 세로 선 생성
+ plt.axhline(y, color)
 - 축을 가로지르는 가로 선 생성

```python
plt.figure(figsize=(10,6))

plt.title('3월 기온')
plt.xlabel('날짜')
plt.ylabel('기온')

plt.plot(temp, 'go--')
plt.grid(True, color ='gray', alpha=0.5, linestyle='--')
plt.axvline(x=10, color='r', linewidth=1)
plt.show()
```

### 글자 쓰기
+ plt.text(x, y, s, fontsize)
 - 원하는 위치에 텍스트 생성

```python
plt.figure(figsize=(10,6))

plt.title('3월 기온')
plt.xlabel('날짜')
plt.ylabel('기온')

plt.plot(temp, 'go--')
plt.grid(True, color ='gray', alpha=0.5, linestyle='--')
plt.axvline(x=10, color='r', linewidth=1)
for i in range(len(temp)) :
  plt.text(i, temp[i]+0.5, temp[i])
plt.show()
```

### x축 눈금 표시 및 회전

```python
avg = [4.5,6.6,10.5,13.4,19.3,21.1,24.3,27,23.1,18.6,13,7]
xt = list(range(1,13))
xt = [str(i) + '월' for i in xt]
plt.figure(figsize=(10,6))
plt.plot(avg, 'go--')
plt.title('기온 변화')
plt.xticks(range(len(avg)), labels=xt, rotation=90)
plt.show()
```

### 범례 표시

```python
plt.figure(figsize=(10,6))
plt.plot(temp, 'g--', label='기본')
plt.plot(temp, 'b--', drawstyle='steps-post', label='계단식')
plt.legend(loc='best')
plt.show()
```

### subplot
+ 하나의 figure에 여러개 그래프 그리기

```python
plt.figure(figsize=(10,4))
plt.subplot(1,2,1) #1행에 2개에 첫번쨰
plt.plot(temp, 'go--')
plt.title('3월 기온1')

plt.subplot(1,2,2)
plt.plot(temp, 'bx')
plt.title('3월 기온2')

plt.show()
```

```python
fig = plt.figure(figsize=(10,4))
ax1= fig.add_subplot(1,2,1)
ax2= fig.add_subplot(1,2,2)
ax1.plot(temp, 'go--')
ax1.set_title('3월 기온1')

ax2.plot(temp, 'bx')
ax2.set_title('3월 기온2')

plt.show()
```

### 다양한 그래프 그리기
+ bar : 막대그래프
+ scatter : 산점도 그래프
+ pie : 파이 그래프
+ hist : 히스토그램 그래프
+ boxplot : 박스플롯 그래프
+ imshow() :  이미지 데이터처럼 행과 열을 가진 행렬 형태의 2차원 데이터 그래프

```python
x=list(range(1,32))
plt.figure(figsize=(10,6))
plt.subplot(2,3,1)
plt.bar(x, temp)
plt.title('막대그래프')

plt.subplot(2,3,2)
plt.barh(x, temp)
plt.title('막대그래프')

plt.subplot(2,3,3)
plt.scatter(x, temp)
plt.title('산포도그래프')

plt.subplot(2,3,4)
plt.pie(temp)
plt.title('파이그래프')

plt.subplot(2,3,5)
plt.hist(temp)
plt.title('히스토그램 그래프')

plt.subplot(2,3,6)
plt.boxplot(temp)
plt.title('박스플롯 그래프')

plt.show()
```
